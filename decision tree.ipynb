```python
import pandas as pd
import numpy as np
from IPython.display import display_html

def display_dfs(dfs, gap=50, justify='center'):
    html = ""
    for title, df in dfs.items():  
        df_html = df._repr_html_()
        cur_html = f'<div> <h3>{title}</h3> {df_html}</div>'
        html +=  cur_html
    html= f"""
    <div style="display:flex; gap:{gap}px; justify-content:{justify};">
        {html}
    </div>
    """
    display_html(html, raw=True)
```

```python
outlook = ['sunny','sunny','overcast','rainy','rainy','rainy','overcast','sunny','sunny','rainy','sunny','overcast','overcast','rainy']
temp = ['hot','hot','hot','mild','cool','cool','cool','mild','cool','mild','mild','mild','hot','mild']
humidity = ['high','high','high','high','normal','normal','normal','high','normal','normal','normal','high','normal','high']
windy = ['FALSE','TRUE','FALSE','FALSE','FALSE','TRUE','TRUE','FALSE','FALSE','FALSE','TRUE','TRUE','FALSE','TRUE']
play = ['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no']
```

```python
df = pd.DataFrame({'outlook':outlook,'temp':temp,'humidity':humidity,'windy':windy,'play':play})
```

```python
df
```

## Outlook

$E(s) = \left(1-\left(P_y^2+P_N^2\right)\right)$

```python
dfs = {'sunny':df[df['outlook']=='sunny'],'overcast':df[df['outlook']=='overcast'],'rainy':df[df['outlook']=='rainy'] }
display_dfs(dfs,justify='flex-start')
gini_outlook = (5/14)*(1-(2**2/25)-(3**2/25)) + (4/14)*(1-(5**2/25)-(0**2/25)) + (5/14)*(1-(3**2/25)-(2**2/25))
print("gini_outlook :",gini_outlook)
```

## Temperature

```python
dfs = {'hot':  df[df['temp']=='hot'], 'mild': df[df['temp']=='mild'], 'cool':  df[df['temp']=='cool']}
display_dfs(dfs, justify='flex-start') 
gini_temp = (4/14)*(1-(2**2/16)-(2**2/16)) + (6/14)*( 1-(4**2/36)-(2**2/36)) + (4/14)*(1-(3**2/16)-(1**2/16))
print('gini_temp',gini_temp)
```

## Humidity

```python
dfs = {'high':df[df['humidity']=='high'],'normal':df[df['humidity']=='normal']}
display_dfs(dfs,justify='flex-start')
gini_humidity = (7/14)*(1-(3**2/49)-(4**2/49)) + (7/14)*(1-(6**2/49)-(1**2/49))
print("gini_humidity :",gini_humidity)
```

## Wind

```python
dfs={'TRUE':df[df['windy']=='TRUE'],'FALSE':df[df['windy']=='FALSE']}
display_dfs(dfs,justify='flex-centre')
gini_windy = (6/14)*(1-(3**2/36)-(3**2/36)) + (8/14)*(1-(6**2/64)-(2**2/64))
print("gini_windy :",gini_windy)
```

## Gini index for columns

```python
feature = ['outlook','temp','humidity','windy']
gini_index = [gini_outlook,gini_temp,gini_humidity,gini_windy]
inspection = pd.DataFrame({'features':feature,'gini_index':gini_index})
inspection
```

##### From table, you can seen that Gini index for outlook feature is lowest. So we get our root node.

### Temperature on sunny outlook

```python
dfs = {'sunny_temp_hot':df[(df['outlook']=='sunny') & (df['temp']=='hot') ],"sunny_temp_cool":df[(df['outlook']=='sunny') & (df['temp']=='cool') ],"sunny_temp_mild":df[(df['outlook']=='sunny') & (df['temp']=='mild') ]}
display_dfs(dfs, justify='flex-centre')
gini_sunny_temp = (2/5)*(1-(0**2/4)-(2**2/4)) + (1/5)*(1-(1**2/1)-(0**2/1)) + (2/5)*(1-(1**2/4)-(1**2/4))
print("gini_sunny_temp:",gini_sunny_temp)
```

### Humidity on sunny outlook

```python
dfs = {'sunny_humidity_high':df[(df['outlook']=='sunny') & (df['humidity']=='high') ],"sunny_humidity_normal":df[(df['outlook']=='sunny') & (df['humidity']=='normal') ]}
display_dfs(dfs, justify='flex-centre')
gini_sunny_humidity = (3/5)*(1-(0**2/9)-(3**2/9)) + (2/5)*(1-(2**2/4)-(0**2/4))
print("gini_sunny_humidity:",gini_sunny_humidity)
```

### Wind on sunny outlook

```python
dfs = {'sunny_windy_false':df[(df['outlook']=='sunny') & (df['windy']=='FALSE') ],"sunny_windy_true":df[(df['outlook']=='sunny') & (df['windy']=='TRUE') ]}
display_dfs(dfs, justify='flex-centre')
gini_sunny_windy = (3/5)*(1-(1/3)**2-(2/3)**2) + (2/5)*(1-(1/2)**2-(1/2)**2)
print("gini_sunny_humidity:",gini_sunny_windy)
```

```python
feature = ['temp','humidity','windy']
gini_index = [gini_sunny_temp,gini_sunny_humidity,gini_sunny_windy]
inspection = pd.DataFrame({'features':feature,'gini_index':gini_index})
inspection
```

#### we have calculated the Gini index of all the features when the outlook is sunny. You can infer that humidity has lowest value. so next node will be humidity.
